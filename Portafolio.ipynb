<h3 Style = 'color:#469272'>4. Manejo de valores faltantes</h3>

<p>Que se hace para el manejo de valores faltantes?</p>
<ol>
    <li>Identificar como se muestran los valores faltantes</li>
    <li>Contar la cantidad de valores faltantes en el set de datos</li>
    <li>Imputación de valores numéricos</li>
    <li>Imputación de valores Categóricos</li>
    <li>Eliminar filas</li>
</ol>

<h5 Style = 'color:#469272'>4.1 Identificar valores faltantes</h5>


```python
dfAutos.replace('?',np.nan,inplace= True)
dfAutos.head(5)
```

<h5 Style = 'color:#469272'>4.2 Contar la cantidad de valores faltantes</h5>

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<h5 Style = 'color:#469272'>4.3 Imputación de valores numéricos</h5>

```python
av_normalized = dfAutos["normalized"].astype("float").mean(axis=0)
print('La Media es: ',av_normalized)
```

```python
dfAutos["normalized"].replace(np.nan,122,inplace=True)
dfAutos.head(5)
```

```python
av_peak = dfAutos["peak"].astype("float").mean(axis=0)
print('La Media es: ',av_peak)
```

```python
dfAutos["peak"].replace(np.nan,5101,inplace=True)
dfAutos.head(5)
```

```python
av_horsePower = dfAutos["horsePower"].astype("float").mean(axis=0)
print('La Media es: ',av_horsePower)
```

```python
dfAutos["horsePower"].replace(np.nan,103,inplace=True)
dfAutos.head(5)
```

```python
av_stroke = dfAutos["stroke"].astype("float").mean(axis=0)
print('La Media es: ',av_stroke)
```

```python
dfAutos["stroke"].replace(np.nan,3,inplace=True)
dfAutos.head(5)
```

```python
av_bore = dfAutos["bore"].astype("float").mean(axis=0)
print('La Media es: ',round(av_bore))
dfAutos["bore"].replace(np.nan,av_bore,inplace=True)
dfAutos.head(5)
```

<h5 Style = 'color:#469272'>5.2 Convertir valores numéricos, en función de la interpretación del medio</h5>

```python
dfAutos["City L/100km (fx)"] = 235/dfAutos["City"]
dfAutos["Higway L/100km (fx)"] = 235/dfAutos["Higway"]
dfAutos.head(5)
```

```python
dfTemp2 = dfAutos[["City L/100km (fx)","Higway L/100km (fx)","engineSize"]]
dfTemp2.head(5)
```

```python
dfAutos["City L/100km (norm)"] = dfAutos["City L/100km (fx)"]/dfAutos["City L/100km (fx)"].max()
dfAutos["Higway L/100km (norm)"] = dfAutos["Higway L/100km (fx)"]/dfAutos["Higway L/100km (fx)"].max()
dfAutos["engineSize (norm)"] = dfAutos["engineSize"]/dfAutos["engineSize"].max()

dfTemp2= dfAutos[["City L/100km (fx)","Higway L/100km (fx)","engineSize","City L/100km (norm)","Higway L/100km (norm)","engineSize (norm)"]]

dfTemp2.head(5)
```

```python
#Remove all axis
mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = False
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
gf = dfAutos.groupby(['make']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling',ascending = False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,6.9), dpi=100)

 

eje_x = gf.index.tolist()
data = gf['symboling'].tolist()

 

bar_colors = ['#89abb4','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
              '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
              '#e5e5e5','#fdd1b6']
ax.barh(eje_x, data, color=bar_colors)

 

#--------------------------------------------------------------------------------------------------------------------#
#--------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GRAFICO ------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO
title = "Grafico 1: Vehiculos"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif")
             #fontname = "Liberation Sans Narrow"

 

#FORMATO DEL SUBTITULO
subTitle = "Cantidad de vehiculos agrupados según fabricante"
ax.set_title(subTitle,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

 

#FORMATO DE LOS EJES
etiqueta = "Cantidad de vehiculos (ABS)"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

 

#CONFIGURACION DE COORDENADAS DE LA ETIQUETA DEL EJE X
ax.xaxis.set_label_coords(x=0,
                          y=-0.07,
                          transform=ax.transAxes)

 

#CONFIGURACION EL TAMANO DEL TEXTO DEL EJE Y
ax.tick_params(axis='both',
               labelsize=10)

 

#CONFIGURACION LA UBICACION DEL GRAFICO
plt.subplots_adjust(top=0.80)
plt.show()

```
